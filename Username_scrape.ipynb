```python
import requests
from bs4 import BeautifulSoup
import json

# Replace this URL with the actual URL of the marketplace page you want to scrape
url = "https://www.vinted.com/catalog?search_text=Jacket&time=1701378132&page=2"

# Send an HTTP request to the URL
response = requests.get(url)
```

```python
html = response.content.decode('utf-8')
```

```python
import re
# Define a regular expression pattern
pattern = r'"login":"([^"]+)"'

# Use re.findall to find all matches
matches = re.findall(pattern, html)

# Print the matches
print(set(matches))
print(len(set(matches)))
```
