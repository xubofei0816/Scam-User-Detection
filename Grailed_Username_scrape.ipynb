```python
import requests
from bs4 import BeautifulSoup
import json
```

```python
#change the url according to your desired search
depop_url = "https://webapi.depop.com/api/v2/search/products/?what=carhartt+jacket&cursor=M3w0OHwxNzAxNDAwMjk5&itemsPerPage=200&country=us&currency=USD&sort=newlyListed"

# Send an HTTP request to the URL
response = requests.get(depop_url)

html = response.content.decode('utf-8')
```

```python
print(html)
```

```python
import re
# Define a regular expression pattern
pattern = r'"slug":"([^"-]+)-'

# Use re.findall to find all matches
matches = re.findall(pattern, html)

# Print the matches
print(set(matches))
print(len(set(matches)))
```

```python
# Save the matches to a text file
with open('freshly_extracted_users.txt', 'w') as file:
    for username in set(matches):
        file.write(username + '\n')

```
